```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# import dataset 
from sklearn.datasets import fetch_california_housing
df=fetch_california_housing()
df
```

```python
df.data
```

```python
# making dataframe
dataset=pd.DataFrame(df.data)
# naming the columns from the feature array of df
dataset.columns=df.feature_names
```

```python
df.feature_names
```

```python
dataset
```

```python
# adding the target variable column and renaming as price
dataset["Price"]=df.target
dataset["Price"]
```

```python
dataset
```

```python
x=dataset.iloc[:,:-1] # independent feature
y=dataset.iloc[:,-1] # dependent feature
```

```python
x.shape
```

```python
y.shape
```

```python
y
```

```python
x
```

```python
y=y.to_numpy()
```

```python
x=x.to_numpy()
```

```python
y=y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lin_regressor=LinearRegression()
```

```python
lin_regressor.fit(x_train,y_train)
```

```python
y_pred_linear=lin_regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
# error calculation
mae=mean_absolute_error(y_test,y_pred_linear)
mse=mean_squared_error(y_test,y_pred_linear)
rmse=np.sqrt(mse) # finds the root mean square error
print(mae)
print(mse)
print(rmse)
```

```python
lin_regressor.coef_
```

```python
print(lin_regressor.score(x_test,y_test))
print(lin_regressor.score(x_train,y_train))
```

```python
from sklearn.linear_model import Lasso
lasso_regressor=Lasso(alpha=0.5)
lasso_regressor.fit(x_train,y_train)
lasso_regressor.predict(x_test)
```

```python
lasso_regressor.coef_
```

```python
from sklearn.model_selection import GridSearchCV
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
grid_search=GridSearchCV(Lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor=grid_search.estimator()
```

```python
lasso_regressor
```

```python
lasso_regressor.best_estimator
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(x_train,y_train)
```
