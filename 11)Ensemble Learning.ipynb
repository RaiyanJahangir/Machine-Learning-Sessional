```python
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv('/diabetes.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum() # no null values
```

```python
x=df.drop(columns=['Outcome'])
y=df['Outcome']
y=y.values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
params_knn={'n_neighbors': np.arange(1,25)}
knn_gs=GridSearchCV(knn,params_knn,cv=5)
knn_gs.fit(x_train,y_train)
```

```python
knn_best=knn_gs.best_estimator_
```

```python
print(knn_gs.best_params_)
```

```python
params_knn
```

```python
knn_gs.predict(x_test)
```

```python
print('knn: {}'.format(knn_best.score(x_test,y_test)))
#print('rf: {}'.format(rf_best.score(x_test,y_test)))
print('dt: {}'.format(dt.score(x_test,y_test)))
print('log_reg: {}'.format(log_reg.score(x_test,y_test)))
```

```python
from sklearn.ensemble import VotingClassifier
estimators=[('knn',knn_best),('dt',dt),('log_reg',log_reg)]
ensemble=VotingClassifier(estimators,voting='hard')
```

```python
ensemble.fit(x_train,y_train)
ensemble.score(x_test,y_test)
```

```python
from sklearn import model_selection
kfold=model_selection.KFold(n_splits=10)
ensemble_cross_val=model_selection.cross_val_score(ensemble,x,y,cv=kfold)
```

```python
print(ensemble_cross_val.mean())
```
